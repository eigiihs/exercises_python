1. Crie uma rotina de calculadora, onde receba dois valores e indique e receba qual operação básica deseja calcular, em seguida apresente o resultado. Todo o cálculo deve ser feito com uso de funções.

```python
def soma(a,b):
    return a + b

def subtrai(a,b):
    return a - b

def multiplica(a,b):
    return a * b

def divide(a,b):
    if b == 0:
        print('Não é possível dividir por zero!')
    return a / b

def get_values():
    a = float(input(f'Informe o 1° valor: '))
    b = float(input(f'Informe o 2° valor: '))
    return a, b

def init():
    running = True
    while running:
        print(f'\n--- Informe o que deseja fazer ---')
        print("""
1- Somar
2- Subtrair
3- Multiplicar
4- Dividir
0- Fechar
""")
        option = int(input())
        if option == 0:
            running = False
            print('Programa encerrado!')
        else: 
            a, b = get_values()
            match option:
                case 1:
                    result = soma(a,b)
                    print(f'O resultado da soma é: {result}')
                case 2:
                    result = subtrai(a,b)
                    print(f'O resultado da subtração é: {result}')
                case 3:
                    result = multiplica(a,b)
                    print(f'O resultado da multiplicação é: {result}')
                case 4:
                    result = divide(a,b)
                    print(f'O resultado da divisão é: {result}')
                case _:
                    print('Opção inválida! Tente novamente.')

init()
```

5. Organiza número. Faça uma rotina que organize os números recebidos em: crescente, decrescente e reverso. Por exemplo:
- recebido: 293
- crescente: 239
- decrescente: 932
- reverso: 392

```python
def crescente(n):
    # transformamos o valor em string para manipular cada dígito, ordenamos e unimos os valores novamente 
    return int(''.join(sorted(str(n))))

def decrescente(n):
    return int(''.join(sorted(str(n), reverse=True)))

def reverso(n):
    # percorrendo a string de trás para frente usando [::,-1]
    return int(str(n)[::-1])

def init():
    num = int(input('Informe um número: '))
    valor_crescente = crescente(num)
    valor_decrescente = decrescente(num)
    valor_reverso = reverso(num)

    print(f'Número recebido: {num}')
    print(f'Crescente: {valor_crescente}')
    print(f'Decrescente: {valor_decrescente}')
    print(f'NReverso: {valor_reverso}')

init()
```

6. Faça um programa que converta da notação de 24 horas para a notação de 12 horas. Por exemplo, o programa deve converter 14:25 em 2:25 P.M. A entrada é dada em dois inteiros. Deve haver pelo menos duas funções: uma para fazer a conversão e uma para a saída. Registre a informação A.M./P.M. como um valor ‘A’ para A.M. e ‘P’ para P.M. Assim, a função para efetuar as conversões terá um parâmetro formal para registrar se é A.M. ou P.M. Inclua um loop que permita que o usuário repita esse cálculo para novos valores de entrada todas as vezes que desejar.

```python
def converte_hrs(hora):
    if hora == 0:
        hora_12 = 12
        periodo = 'A'
    elif hora == 12:
        hora_12 = 12
        periodo = 'P'
    elif hora > 12:
        hora_12 = hora - 12
        periodo = 'P'
    else:
        hora_12 = hora 
        periodo = 'A'

    return hora_12, periodo

def exibe_hrs(hora_12, minutos, periodo):
    print(f'{hora_12}:{minutos} - {periodo}.M.')

def menu_inicial():
    print('\nSelecione uma opção: ')
    print('1- Converter novo horário')
    print('0- Sair')

def init():
    print('--- Iniciando Sistema de Conversão de Horas ---')
    while True:
        menu_inicial()

        option = int(input('Informe a opção (1/0): '))

        if option == 1: 
            try: 
                hora_24 = int(input('Informe a hora (formato 24 horas): '))
                if hora_24 < 0 or hora_24 > 23:
                    print('Hora inválida! Informe novamente.')
                    continue
            except:
                print('Entrada inválida! Informe novamente um número entre 0 e 23.')
                continue

            try:
                minutos = int(input('Informe os minutos: '))
                if minutos < 0 or minutos > 59:
                    print('Minutos inválidos! Informe novamente.')
                    continue
            except:
                print('Entrada inválida! Informe novamente um número entre 0 e 59.')
                continue

            hora_12, periodo = converte_hrs(hora_24)

            exibe_hrs(hora_12, minutos, periodo)
        elif option == 0:
            print('Programa encerrado!')
            break 
        else: 
            print('Opção Inválida! Tente novamente.')

init()
```

7. Data com mês por extenso. Construa uma função que receba uma data no formato DD/MM/AAAA e devolva uma string no formato D de mesPorExtenso de AAAA. Opcionalmente, valide a data e retorne NULL caso a data seja inválida.

```python
def valida_data(dia, mes, ano):
    if mes < 1 or mes > 12:
        print('Mês inválido!')
        return False
    if ano < 1000 or ano > 9999:
        print('Ano inválido! Informe um ano com 4 dígitos.')
        return False
    
    meses_31 = [1, 3, 5, 7, 8, 10, 12]
    meses_30 = [4, 6, 9, 11]
    
    # validando se o ano é bissexto
    # - divisível por 4
    # - não divisível por 100
    # - pode ser divisível por 400
    def bissexto():
        return (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)
    
    if mes in meses_31:
        return (dia >= 1) and (dia <= 31)
    elif mes in meses_30:
        return (dia >= 1) and (dia <= 30)
    elif mes == 2:
        # Fevereiro: se for bissexto, o mês pode ter até 29 dias
        if bissexto():
            return (dia >= 1) and (dia <= 29)
        else:
            return (dia >= 1) and (dia <= 28)
    return False
    
def data_extenso(data):
    meses = { 1:'janeiro', 2:'fevereiro', 3:'março', 4:'abril', 5:'maio', 6:'junho', 7:'julho', 8:'agosto', 9:'setembro', 
        10:'outubro', 11:'novembro', 12:'dezembro'}
    
    # Separa a data em dia, mês e ano
    dia, mes, ano = map(int, data.split('/'))

    # Valida a data e retorna NULL caso seja inválida
    if not valida_data(dia, mes, ano):
        return "NULL"  

    # Converte a data no formato solicitado
    return f"{dia} de {meses[mes]} de {ano}"  

def menu_inicial():
    print('\nSelecione uma opção: ')
    print('1- Inserir nova data')
    print('0- Sair')

def init():
    print('--- Iniciando Sistema de Conversão de Horas ---')
    while True:
        menu_inicial()

        option = int(input('Informe a opção (1/0): '))

        if option == 1:
            data = input('Informe a data no formato DD/MM/AAAA: ')

            result = data_extenso(data)
            print(result)
        elif option == 0:
            print('Programa encerrado!')
            break
        else:
            print('Opção inválida! Tente novamente.')

init()
```
